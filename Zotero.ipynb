```python
%matplotlib inline
import matplotlib.pyplot as plt
from collections import Counter
import requests

library_id = "655084"
library_type = "groups"
api_key= "PASTE KEY HERE"
url = "https://api.zotero.org/groups/"+library_id+"/collections/"+"95HW46ZG"+"/items/top?v=3+key="+api_key
url
```

```python
import json
r = requests.get(url)
data = json.loads(r.text)
```

```python
dates = [int(i["data"]["date"]) for i in data]

years = [d for d in set(dates)]
years.sort()
counts = Counter([u for u in dates])
articles = [counts[y] for y in years]

plt.figure(figsize=(20,10))
plt.plot(years[4:], articles[4:])
plt.show()
```
